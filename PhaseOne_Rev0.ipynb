```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
 
wnl = WordNetLemmatizer()
```

```python
import nltk
nltk.download('omw-1.4')
```

```python
def tokenize(s):
    out=[]
    lm_out=[]
    for i in s:
      out.extend(i.split())
    for words in out:
       lm_out.append(wnl.lemmatize(words))
    return lm_out
```

```python
from scipy import spatial
import gensim.downloader as api
import numpy as np
# Word2Vec

model = api.load("glove-wiki-gigaword-50") #choose from multiple models https://github.com/RaRe-Technologies/gensim-data
```

```python
def model_word(x):
    try:
        c = model[x]
    except KeyError:
        c = 0
        
    return c  
```

```python
model_word('gladiator')
```

```python
# Step 5
df_tool_chest_combos['Tokenized Vector'] = df_tool_chest_combos['Tokenized'].apply(lambda x: np.sum(np.array([model_word(i.lower()) for i in x]), axis=0))
```

```python
df_tool_chest_combos
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
max_scores = []
max_indices = []

for i in range(len(df_tool_chest_combos)):
  # i == 0
  
  sim_scores = []
  for j in range(len(df_tool_chest_combos)):
    # j == 0 to 10

      if(i == j):
        sim_scores.append(-1)
        continue

      score = cosine_similarity(df_tool_chest_combos['Tokenized Vector'][i].reshape(1, -1), df_tool_chest_combos['Tokenized Vector'][j].reshape(1, -1))
      sim_scores.append(score[0][0])

  max_index = np.argmax(sim_scores)
  max_scores.append(sim_scores[max_index])
  max_indices.append(max_index)
```

```python
df_portable_band_saws = df_portable_band_saws[['id','product_title', 'manufacturing_brand', 'product_url']]
df_portable_band_saws.head()
```

```python
df_portable_band_saws['Product_Title_Cleaned'] = df_portable_band_saws['product_title'].str.replace('&','').str.replace('-','').str.replace('.', '').str.replace(',', '').str.replace("'", '').str.split(' ')
```

```python
df_portable_band_saws['Tokenized'] = df_portable_band_saws['Product_Title_Cleaned'].apply(lambda x: tokenize(x))
```

```python
df_portable_band_saws['Tokenized Vector'] = df_portable_band_saws['Tokenized'].apply(lambda x: np.sum(np.array([model_word(i.lower()) for i in x]), axis=0))
```

```python
df_portable_band_saws.reset_index(inplace = True)
```

```python
max_scores = []
max_indices = []

for i in range(len(df_portable_band_saws)):
  
  sim_scores = []
  for j in range(len(df_portable_band_saws)):

      if(i == j):
        sim_scores.append(-1)
        continue

      score = cosine_similarity(df_portable_band_saws['Tokenized Vector'][i].reshape(1, -1), df_portable_band_saws['Tokenized Vector'][j].reshape(1, -1))
      sim_scores.append(score[0][0])

  max_index = np.argmax(sim_scores)
  max_scores.append(sim_scores[max_index])
  max_indices.append(max_index)
```

```python
data = {
  'Product Id': df_portable_band_saws['id'],
  'Product URL': df_portable_band_saws['product_url'],
  'Product Title': df_portable_band_saws['product_title'],
  'Most Similar Product': [df_portable_band_saws['product_title'][i] for i in max_indices],
  'Similarity Score': max_scores,
  'Product Id of most similar product': [df_portable_band_saws['id'][i] for i in max_indices],
  'Product URLof most similar product': [df_portable_band_saws['product_url'][i] for i in max_indices]
}

df_portable_band_saws_most_sim = pd.DataFrame(data)
```
